## drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)   # drop
```

```python
print(df.head())
```

# Change weekly study hours column

```python
 df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05 - Oct", "5 - 10")
df.head()
```

## Gender Distribution


```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# from the above chart we have analyzed that:
# the number of females in data is more than mens
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's education and student's score")
plt.show()
```

```python
# from the above chart we have concluded that the education of parents has good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's marital status and student's score")
plt.show()
```

```python
# from the above chart we have concluded that the marital status of parents has negligible impact on their scores
```

## distribution of EthnicGroup

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]

mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distibution of Ethnic Group")
plt.show()
```
